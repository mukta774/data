```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_california_housing
from sklearn.preprocessing import StandardScaler

data = fetch_california_housing()
X = pd.DataFrame(data.data, columns=data.feature_names)
y = pd.DataFrame(data.target, columns=["PRICE"])
```

```python
scaler = StandardScaler()
X_scaled =  pd.DataFrame(scaler.fit_transform(X),columns=data.feature_names)
print(X_scaled)
```

```python
X_scaled = np.c_[np.ones(X_scaled.shape[0]), X_scaled] # This makes a matrix with a column full of ones , np.c_ adds the column of one
y_scaled  = y.values # converts the column into arrays
```

```python
theta = np.zeros((X_scaled.shape[1], 1)) #This creates a NumPy array called theta filled with zeros, The shape of this array is determined by the number of features in X_scaled (which includes the added column of ones) and 1,
# representing the weights (or coefficients) for each feature in the linear regression model. These weights will be updated during the training process.
alpha = 0.01 # learning rate
iterations = 1000
```

```python
def compute_cost(X, y, theta):
    m = len(y)
    predictions = np.dot(X,theta) # performs the dot product of X with theta
    error = predictions - y # loss
    cost = (1 / (2 * m)) * np.dot(error.T, error) # this gives np.dot(error.T, error) as 1*1 array

    return cost[0][0]
```

```python
def gradient_descent(X, y, theta, alpha, iterations):
    m = len(y)
    for _ in range(iterations):
        predictions = X.dot(theta)
        error = predictions - y
        gradient = (1 / m) * X.T.dot(error)  # partial derivatives
        theta = theta - alpha * gradient     # update rule
    return theta
```

```python
def predict(X, theta):
  result = X.dot(theta)
  return result
```

```python
theta = gradient_descent(X_scaled, y_scaled, theta, alpha, iterations)
compute_cost(X_scaled, y_scaled, theta)
```

```python
user_input = input("Enter the MedInc, HouseAge, AveRooms, AveBedrms, Population, AveOccup, Latitude, Longitude (comma-separated values): ")
values = [float(x) for x in user_input.split(',')]

scaled_input = scaler.transform([values])
X_input = np.c_[np.ones((scaled_input.shape[0], 1)), scaled_input]

predicted_price = predict(X_input, theta)
print("Predicted Price:", predicted_price)
```
