```python
import numpy as np

def sigmoid(z):
    return 1 / (1 + np.exp(-z))
```

```python
def initialize_weights(n_features):
    theta = np.zeros((n_features, 1))  # Shape: (n_features, 1)
    return theta
```

```python
def compute_loss(y, y_pred):
    m = y.shape[0]
    loss = - (1/m) * np.sum(y * np.log(y_pred) + (1 - y) * np.log(1 - y_pred))
    return loss
```

```python
def gradient_descent(X, y, theta, alpha, iterations):
    m = X.shape[0]
    for i in range(iterations):
        z = X.dot(theta)
        y_pred = sigmoid(z)
        gradient = (1/m) * X.T.dot(y_pred - y)
        theta = theta - alpha * gradient

        if i % 100 == 0:
            loss = compute_loss(y, y_pred)
            print(f"Iteration {i}: Loss = {loss:.4f}")
    return theta
```

```python
def predict(X, theta, threshold=0.5):
    probs = sigmoid(X.dot(theta))
    return (probs >= threshold).astype(int)
```

```python
# Generate dummy data for testing
np.random.seed(0)
X_dummy = np.random.randn(100, 2)
X_dummy = np.c_[np.ones((X_dummy.shape[0], 1)), X_dummy]  # Add bias term
y_dummy = (X_dummy[:, 1] + X_dummy[:, 2] > 0).astype(int).reshape(-1, 1)

theta = initialize_weights(X_dummy.shape[1])
theta = gradient_descent(X_dummy, y_dummy, theta, alpha=0.1, iterations=1000)

y_pred = predict(X_dummy, theta)
accuracy = np.mean(y_pred == y_dummy)
print(f"Training Accuracy: {accuracy * 100:.2f}%")
```
